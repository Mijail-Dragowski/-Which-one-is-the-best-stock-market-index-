# Return

### Daily return calculation


```python
SP500["daily return"]=SP500["Adj Close"].pct_change(1)
Nasdaq["daily return"]=Nasdaq["Adj Close"].pct_change(1)
DowJones["daily return"]=DowJones["Adj Close"].pct_change(1)
FTSE["daily return"]=FTSE["Adj Close"].pct_change(1)
DAX["daily return"]=DAX["Adj Close"].pct_change(1)
Nikkei["daily return"]=Nikkei["Adj Close"].pct_change(1)
CAC["daily return"]=CAC["Adj Close"].pct_change(1)
Euro_zone["daily return"]=Euro_zone["Adj Close"].pct_change(1)
RUSSIA["daily return"]=RUSSIA["Adj Close"].pct_change(1)
Kospi["daily return"]=Kospi["Adj Close"].pct_change(1)
HANG_SENG["daily return"]=HANG_SENG["Adj Close"].pct_change(1)

```

### Accumulated daily return calculation

```python
SP500['Cumulative Return'] = (1 + SP500['daily return']).cumprod()
Nasdaq['Cumulative Return'] = (1 + Nasdaq['daily return']).cumprod()
DowJones['Cumulative Return'] = (1 + DowJones['daily return']).cumprod()
FTSE['Cumulative Return'] = (1 + FTSE['daily return']).cumprod()
DAX['Cumulative Return'] = (1 + DAX['daily return']).cumprod()
Nikkei['Cumulative Return'] = (1 + Nikkei['daily return']).cumprod()
CAC['Cumulative Return'] = (1 + CAC['daily return']).cumprod()
Euro_zone['Cumulative Return'] = (1 + Euro_zone['daily return']).cumprod()
RUSSIA['Cumulative Return'] = (1 + RUSSIA['daily return']).cumprod()
Kospi['Cumulative Return'] = (1 + Kospi['daily return']).cumprod()
HANG_SENG['Cumulative Return'] = (1 + HANG_SENG['daily return']).cumprod()
```

### Accumulated daily return time series chart


```python
SP500['Cumulative Return'].plot(label='SP500',figsize=(16,8),title='Cumulative Return')
Nasdaq['Cumulative Return'].plot(label='Nasdaq')
DowJones['Cumulative Return'].plot(label='Dow Jones')
FTSE['Cumulative Return'].plot(label='FTSE')
DAX['Cumulative Return'].plot(label='DAX')
Nikkei['Cumulative Return'].plot(label='Nikkei 225')
CAC['Cumulative Return'].plot(label='CAC 40')
Euro_zone['Cumulative Return'].plot(label='Euro zone')
RUSSIA['Cumulative Return'].plot(label='Russia')
Kospi['Cumulative Return'].plot(label='Kospi')
HANG_SENG['Cumulative Return'].plot(label='HANG_SENG')
plt.legend()

```

# Sharpe ratio

### Annualized sharpe ratio calculation


```python
ASR_SP500 = ((SP500['daily return'].mean()*252)-0.0498)/(SP500['daily return'].std()*252**0.5)
ASR_Nasdaq = ((Nasdaq['daily return'].mean()*252)-0.0498)/(Nasdaq['daily return'].std()*252**0.5)
ASR_DowJones = ((DowJones['daily return'].mean()*252)-0.0498)/(DowJones['daily return'].std()*252**0.5)
ASR_FTSE = ((FTSE['daily return'].mean()*252)-0.0498)/(FTSE['daily return'].std()*252**0.5)
ASR_DAX = ((DAX['daily return'].mean()*252)-0.0498)/(DAX['daily return'].std()*252**0.5)
ASR_Nikkei = ((Nikkei['daily return'].mean()*252)-0.0498)/(Nikkei['daily return'].std()*252**0.5)
ASR_CAC = ((CAC['daily return'].mean()*252)-0.0498)/(CAC['daily return'].std()*252**0.5)
ASR_Euro_zone = ((Euro_zone['daily return'].mean()*252)-0.0498)/(Euro_zone['daily return'].std()*252**0.5)
ASR_RUSSIA = ((RUSSIA['daily return'].mean()*252)-0.0498)/(RUSSIA['daily return'].std()*252**0.5)
ASR_Kospi = ((Kospi['daily return'].mean()*252)-0.0498)/(Kospi['daily return'].std()*252**0.5)
ASR_HANG_SENG = ((HANG_SENG['daily return'].mean()*252)-0.0498)/(HANG_SENG['daily return'].std()*252**0.5)
```

### View of each  annualized sharpe ratio for all index

```python
df=[ASR_SP500, ASR_Nasdaq, ASR_DowJones, ASR_FTSE, ASR_DAX, ASR_Nikkei, ASR_CAC, ASR_Euro_zone, ASR_RUSSIA,ASR_Kospi, ASR_HANG_SENG]
ASR= pd.DataFrame([df])  
ASR.columns=["ASR_SP500", "ASR_Nasdaq", "ASR_DowJones", "ASR_FTSE", "ASR_DAX", "ASR_Nikkei", "ASR_CAC"," ASR_Euro_zone", "ASR_RUSSIA","ASR_Kospi", "ASR_HANG_SENG"]

ASR
```

### Graph of each annualized sharpe ratio

```python
fig, ax = plt.subplots(figsize=(8, 6))
ASR.plot.bar(ax=ax,ec="black")

ax.set_title('annualized sharpe ratio')
ax.set_xlabel("Index")
ax.set_ylabel('ASR values')
ax.spines['right'].set_visible(False) 
ax.spines['top'].set_visible(False)  
ax.tick_params(axis='x', length=0) 
ax.grid(axis='y', linestyle='--', alpha=0.7)  
ax.legend(loc='upper right', frameon=False, bbox_to_anchor=(1.15, 1)) 
plt.xticks(rotation=0)  

colors = ['red', 'green', 'blue',"yellow", "pink", "grey","black","orange","purple","brown", "gold"]
for i, c in enumerate(ax.containers):
    
    ax.bar_label(c, label_type='center', fontsize=0, color='black')

ax.set_xticklabels([])
plt.tight_layout()
plt.show()
```
